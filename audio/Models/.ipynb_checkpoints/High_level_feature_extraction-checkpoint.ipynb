# High level features extraction

The goal of this notebook is to extract the following high level features from the MIT dataset: number_ of_syllables, number_of_pauses, rate_of_speech, articulation_rate, speaking_duration, original_duration, balance.
This is done using the libraries Myprosody and Praat. In the end of this notebook a csv file named "high_level_features.csv" is generated and it contains the features for each interview.

## Importing packages

## Extracting high level features from the MIT dataset with Myprosody and Praat

```python
m="P1" # Audio File title
p="/Users/ahmedewva/Desktop/School/3A/APC/code/myprosody" # Path to the Audio_File directory (Python 3.7)
my_dir = "/Users/ahmedewva/Desktop/School/3A/APC/code/myprosody/dataset/audioFiles"
```

```python
def my_features(m,p):
    sound=p+"/"+"dataset"+"/"+"audioFiles"+"/"+m+".wav"
    sourcerun=p+"/"+"dataset"+"/"+"essen"+"/"+"myspsolution.praat"
    path=p+"/"+"dataset"+"/"+"audioFiles"+"/"
    try:
        objects= run_file(sourcerun, -20, 2, 0.3, "yes",sound,path, 80, 400, 0.01, capture_output=True)
        print (objects[0]) # This will print the info from the sound object, and objects[0] is a parselmouth.Sound object
        z1=str( objects[1]) # This will print the info from the textgrid object, and objects[1] is a parselmouth.Data object with a TextGrid inside
        z2=z1.strip().split()
        z3=np.array(z2)
        z4=np.array(z3)[np.newaxis]
        z5=z4.T
        dataset=pd.DataFrame({"number_ of_syllables":z5[0,:],"number_of_pauses":z5[1,:],"rate_of_speech":z5[2,:],"articulation_rate":z5[3,:],"speaking_duration":z5[4,:],
                          "original_duration":z5[5,:],"balance":z5[6,:]})
        return dataset
    except:
        print ("Try again the sound of the audio was not clear")
    return;
```

```python
def high_level_features (files, mydf, target_label, target_column):
    n = len(files)
    df = pd.DataFrame(columns=["Participant", "simple_feature_dict", target_label], dtype=object)
    
    labels_df_raw = pd.read_csv("MIT_dataset_APC/Labels/turker_scores_full_interview.csv")

    for i in range(n):

        labels_audio = ['Worker',target_label]#,'NotAwkward','Excited','SpeakingRate','NoFillers','Friendly','EngagingTone','Calm','NotStressed']
        bis = labels_df_raw.loc[labels_df_raw["Participant"] == files[i].lower() ,]
        bis = bis.reindex(columns = labels_audio)
        
        score = bis.loc[bis["Worker"] == 'AGGR' ,[target_label]].values[0,0]
        
        x = mydf.loc[mydf["Participant"] == files[i] ,[target_column]].values[0,0]
        #speech = mydf.loc[mydf["Participant"] == files[i] ,["rate_of_speech"]].values[0,0]
        #articulation = mydf.loc[mydf["Participant"] == files[i] ,["articulation_rate"]].values[0,0]
        print(x, type(x))
        try:
            x = float(x)
        except ValueError:
            x = 0
            print("Not a float")
        df = df.append({"Participant":files[i], "simple_feature_dict":{"target":x} , target_label:score}, ignore_index=True)

    return df
```

```python
target_label = "RecommendHiring"
df_Test = high_level_features (files_interv, file_df, target_label,"balance")
```

```python
def simple_aggregate(series):
    my_list = []

    out = {
        'target': my_list,
    }
    for dict in series:
        my_list.append(dict.get('target'))
    return out
```

```python
dict_good_simple = df_Test[index_good_interviews]['simple_feature_dict'].aggregate(simple_aggregate)
dict_bad_simple = df_Test[index_bad_interviews]['simple_feature_dict'].aggregate(simple_aggregate)
```
